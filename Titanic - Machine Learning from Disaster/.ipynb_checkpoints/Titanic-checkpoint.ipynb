```python
temp_df = df[df['Survived'] == 0].groupby('Pclass')
temp_df1 = df[df['Survived'] == 1].groupby('Pclass')
array = np.array([[len(temp_df.get_group(1)),len(temp_df.get_group(2)),len(temp_df.get_group(3))],
                  [len(temp_df1.get_group(1)),len(temp_df1.get_group(2)),len(temp_df1.get_group(3))]])

species = ('Class 1', 'Class 2', 'Class 3')
sex_counts = {
    'Survive': array[1],
    'Deseace': array[0],
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    print(f'{sex}---{sex_count}')
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Survival comparation by class')
ax.legend()

plt.show()
```

### Loking at the image above the class was a key factor for surviving

```python
temp_df = df[df['Survived'] == 0].groupby('Sex')
temp_df1 = df[df['Survived'] == 1].groupby('Sex')
array = np.array([[len(temp_df.get_group('male')),len(temp_df.get_group('female'))],
                  [len(temp_df1.get_group('male')),len(temp_df1.get_group('female'))]])

species = ('Male', 'Female')
sex_counts = {
    'Survive': array[1],
    'Deseace': array[0],
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(2)

for sex, sex_count in sex_counts.items():
    print(f'{sex}---{sex_count}')
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center',padding = 0, size =10)

ax.set_title('Survival comparation by sex')
ax.legend()

plt.show()
```
